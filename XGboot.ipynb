```python
import pandas as pd
from sklearn.model_selection import train_test_split

from xgboost import XGBClassifier
```

```python
# Load dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
data = pd.read_csv(url, delimiter=";")
```

```python
data
```

```python
X = data.drop('quality', axis=1)

y = data['quality']

#Normalize y from value range [3 4 5 6 7 8] to [0 1 2 3 4 5]

z = y-3

type(X)
type(y)
print(z.head)
```

```python

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, z, test_size=0.2, random_state=7)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
y_train_set = set(y_train)
print(y_train_set)
```

```python
# Define the model
model = XGBClassifier()

# Fit the model
model.fit(X_train, y_train)
```

```python
# Saving XGBoost Models

model.save_model('xgb_model.json')
```

```python
# Save model using booster
booster = model.get_booster()
booster.save_model('booster_model.json')
```

```python
loaded_model = XGBClassifier()
loaded_model.load_model('xgb_model.json')
```

```python
# Load model using booster
import xgboost

booster = xgboost.Booster()
booster.load_model('booster_model.json')
```

```python
# Make predictions on test data
y_pred = loaded_model.predict(X_test)
```
